```python
num=[1,2,3,4,5]
n=len(num)
num.sort()
if n%2==0:
    m1=[n//2]
    m2=num[(n//2)-1]
    m=(m1+m2)/2
else:
    m=num[n//2]
print(m)
```

```python
from collections import Counter
num=[1,2,3,4,5,5]
n=len(num)

data=Counter(num)
get_mode=dict(data)
mode=[k for k, v in get_mode.items() if v== max(list(data.values()))]

if len(mode)==n:
    get_mode="No mode found"
else:
    get_mode="Mode is: " + ','.join(map(str,mode))
print(get_mode)
```

```python
import pandas as pd
data={
    'Age_Group': ['Young','Young','Middle','Middle','Middle','Senior','Senior'],
    'Income': [25000,32000,45000,38000,55000,48000,60000]
}

df=pd.DataFrame(data)
print(df.dtypes)
print(df)
```

```python
stats = df.groupby('Age_Group').agg(['mean','median','min','max','std'])
print(stats)
```
